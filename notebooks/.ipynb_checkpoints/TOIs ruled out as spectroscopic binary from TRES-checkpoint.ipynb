## TOI-1385
* [qlp-s16-cameras-1-2-ffi](https://docs.google.com/viewer?url=https%3A%2F%2Ftev.mit.edu%2Ffiles%2F194816%2Fdownload%2F)
* see figure: https://groups.google.com/forum/?utm_medium=email&utm_source=footer#!msg/tfop-sg1/9XyKxvIQPKw/ZLCKG1NAAQAJ

TIC 329864959.01 (TOI-1385.01)

John and I observed a full on 20191102 in zp and confirmed a ~30 ppt ~on-time transit using a 4" aperture that is not contaminated with Gaia or TIC stars bright enough to produce the TESS signal. The zp depth appears to be significantly deeper than the T-band QLP depth, but let's get a blue observation to make a final chromaticity call. There's also some gravity darkening, or some other non-symmetrical effect in the bottom of the light curve. This is also apparent in the individual QLP light curves, but seems to occur at different in-transit phases.

SG2/TRES: The first TRES observation reveals a hot star that is rotating too rapidly to allow precise velocities. 

Update:

Turns out this is a known double star from 7 high resolution observations over 30 years, with separation ~0.3", equal magnitudes, and spectral type A0 (consistent with the TRES spectrum). The stellar radius is therefore large and the 3% event leads to a companion radius that is definitely stellar. (2 Rsun implies >30 Rearth companion). Plus, the transit depth is diluted by a factor of ~2 since TIC8 didn't know about the double star.

VPC => EB (and retired)
