## short cadence

### PDCSAP, SAP

```python
%matplotlib inline
import tql

#uses first available sector and PDCSAP lightcurves by default
fig = tql.plot_tql(toiid=125)
```

```python
#specify sector and lightcurve type
fig = tql.plot_tql(toiid=125, sector=2, lctype='sap')
```

### custom aperture shape

```python
#specify sector and lightcurve type
fig = tql.plot_tql(toiid=125, 
                   sector=2, 
                   lctype='custom', 
                   sap_mask='square', 
                   aper_radius=1
                  )
```
